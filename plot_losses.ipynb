```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fp = "results/srgnn_baseline/progress.csv"
```

```python
progress_df = pd.read_csv(fp)
plt.plot(progress_df["train_loss"], label="train loss")
plt.plot(progress_df["val_loss"], label="val loss")
plt.legend()
plt.xlabel("Steps")
plt.ylabel("Loss")
```
